```python
import numpy as np
import pandas as pd
from collections import defaultdict

%matplotlib inline
import seaborn as sns
import matplotlib.pyplot as plt

import plot_tools as pt
from strong_toolset import WeightliftingLog
```

```python
#wl = WeightliftingLog('/home/mmayers/Dropbox/Workout-Stuff/workout-data/05-09-18-Strong.csv')
wl = WeightliftingLog('/Users/mmayers/Dropbox/Workout-Stuff/workout-data/2018-10-17-strong.csv', 'kg')
```

#### Initial Variables

```python
exercises = ['Snatch', 'Clean', 'Jerk', 'Squat']
progress = defaultdict(dict)
cycle_dates = dict()
prs = []
```

## Dave's Programming

```python
cycle = "Dave's 0"

start_date = '2017-01-01'
end_date = '2017-04-02'

cycle_dates[cycle] = {'start': wl._format_date(start_date), 'end': wl._format_date(end_date)}

wl.set_date_range(start_date, end_date, inplace=True)
```

```python
fig, ax = plt.subplots(1, 2)
fig.set_size_inches(14, 4)

pt.plot_category_volume(wl, ax[0])
pt.plot_weekly_workouts(wl, ax[1])
```

```python
fig, ax = plt.subplots(4, 2, sharey='row', sharex=True)
fig.set_size_inches(14, 20)

for i, exercise in enumerate(exercises):
    intensity = wl.get_weekly_intensity(exercise).set_index(['date', 'exercise'])['intensity']
    volume = wl.get_weekly_volume(exercise).set_index(['date', 'exercise'])['volume']
    calc_1rm = wl.get_weekly_1rm(exercise).set_index(['date', 'exercise'])['calc_1rm']


    pt.dual_axis_plot(volume, calc_1rm, wl.get_weight(), wl.get_weight(), ax[i][0])
    pt.dual_axis_plot(volume, intensity, wl.get_weight(), '%', ax[i][1])
```

```python
for exercise in exercises:
    maxes = wl.get_weekly_1rm(exercise)
    start = maxes.iloc[0, 1]
    top = maxes['calc_1rm'].max()
    
    progress[cycle].update({exercise+'_start': start, exercise+'_best': top})
```

```python
cycle_prs = wl.find_cycle_prs(start_date, end_date)
cycle_prs['cycle'] = cycle
prs.append(cycle_prs)
```

# PR results

```python
prs = pd.concat(prs, sort=False).reset_index(drop=True)
```

```python
prs.head(2)
```

```python
prs['cycle'].value_counts()
```

```python
prs.groupby('cycle')['absolute_max'].sum().astype(int).sort_values(ascending=False)
```

```python
the_big_ones = ['Clean', 'Split Jerk', 'Snatch', 'Front Squat', 'Squat', 'Bench Press', 'Deadlift', 'Strict Press']
big_lift_prs = prs.query('exercise in @the_big_ones').copy()

big_lift_prs['cycle'].value_counts()
```

```python
big_lift_prs.groupby('cycle')['absolute_max'].sum().astype(int).sort_values(ascending=False)
```

```python
big_lift_prs.query('absolute_max')
```

```python
big_lift_prs['cycle_start'] = big_lift_prs['cycle'].apply(lambda x: cycle_dates[x]['start'])

melt_prs = big_lift_prs.query('absolute_max').melt(id_vars=['cycle', 'exercise'], 
                                                   value_vars=['previous_max', 'new_max'])
melt_prs.head()
```

```python
start_idx = melt_prs['variable'] == 'previous_max'
end_idx = melt_prs['variable'] == 'new_max'

melt_prs.loc[start_idx, 'date'] = melt_prs.loc[start_idx, 'cycle'].apply(lambda x: cycle_dates[x]['start'])
melt_prs.loc[end_idx, 'date'] = melt_prs.loc[end_idx, 'cycle'].apply(lambda x: cycle_dates[x]['end'])
```

```python
sns.lineplot(x='date', y='value', hue='exercise', data=melt_prs)
plt.title('Achieved new 1RMs by Date')
plt.xlabel('Dates')
plt.xticks(rotation=45)
plt.ylabel('Weight (kg)')
plt.legend(title='', loc='right', bbox_to_anchor=(1.35, .5));
```

## Program Analysis

```python
progress = pd.DataFrame(progress).T
```

```python
progress.head(2)
```

```python
for ex in exercises:
    progress[ex+'_change'] = progress[ex+'_best'] - progress[ex+'_start']
    progress[ex+'_per_inc'] = (progress[ex+'_best'] / progress[ex+'_start']) - 1
```

```python
progress.head(2)
```

```python
# Reformat the data
cols = progress.columns
melt_dat = progress.reset_index().rename(columns={'index': 'cycle'}).melt(id_vars=['cycle'], value_vars=cols)

melt_dat['start_date'] = melt_dat['cycle'].apply(lambda x: cycle_dates[x]['start'])

melt_dat = melt_dat.sort_values(['start_date', 'variable']).reset_index(drop=True)
```

```python
melt_dat.head(2)
```

```python
data = melt_dat[melt_dat['variable'].str.contains('best')]

sns.pointplot(x='cycle', y='value', data=data, hue='variable')
plt.title('Best Calcualted 1RM per Cycle')
plt.xticks(rotation=45)
plt.xlabel('')
plt.ylabel('Weight (kg)')
l = plt.legend(loc='right', bbox_to_anchor=(1.25, .5))
for t in l.get_texts():
    text = t.get_text()
    t.set_text(text.split('_')[0])
```
