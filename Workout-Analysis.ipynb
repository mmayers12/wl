```python
import numpy as np
import pandas as pd
from collections import defaultdict

%matplotlib inline
import seaborn as sns
import matplotlib.pyplot as plt

import plot_tools as pt
from strong_toolset import WeightliftingLog
```

```python
wl = WeightliftingLog('/home/mmayers/Dropbox/Workout-Stuff/workout-data/03-24-2018-Strong.csv')
```

```python
exercises = ['Snatch', 'Clean', 'Jerk', 'Squat']
progress = defaultdict(dict)
```

## Dave's Programming

```python
cycle = "Dave's 0"

start_date = '2017-01-01'
end_date = '2017-04-02'

wl.set_date_range(start_date, end_date)
```

```python
fig, ax = plt.subplots(1, 2)
fig.set_size_inches(14, 4)

pt.plot_category_volume(wl, ax[0])
pt.plot_weekly_workouts(wl, ax[1])
```

```python
fig, ax = plt.subplots(4, 2, sharey='row', sharex=True)
fig.set_size_inches(14, 20)

for i, exercise in enumerate(exercises):
    intensity = wl.get_weekly_intensity(exercise).set_index(['date', 'exercise'])['intensity']
    volume = wl.get_weekly_volume(exercise).set_index(['date', 'exercise'])['volume']
    calc_1rm = wl.get_weekly_1rm(exercise).set_index(['date', 'exercise'])['calc_1rm']


    pt.dual_axis_plot(volume, calc_1rm, wl.get_weight(), wl.get_weight(), ax[i][0])
    pt.dual_axis_plot(volume, intensity, wl.get_weight(), '%', ax[i][1])
```

```python
for exercise in exercises:
    maxes = wl.get_weekly_1rm(exercise)
    start = maxes.iloc[0, 1]
    top = maxes['calc_1rm'].max()
    
    progress[cycle].update({exercise+'_start': start, exercise+'_best': top})
```
